### Introduction:
**We are going to use Apple's stock price.**


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pylab as plt
```

### Step 2. Import the dataset


```python
pd.read_csv('apple.csv')
```


### Step 3. Assign it to a variable apple


    

```python
apple = pd.read_csv('apple.csv')
```

 

### Step 4. Check out the type of the columns



```python
apple.columns
```



     

### Step 5. Transform the Date column as a datetime type

   

```python
apple['Date'] = pd.to_datetime(apple['Date'])
apple['Date']
# apple.columns
```

  



     

### Step 6. Set the date as the index


 

```python
apple.set_index(apple['Date'], inplace=True)
apple
```

    

### Step 7. Is there any duplicate dates?



```python
du = apple.index.duplicated()
if any(du) :
    print("Duplicate date found")
else :
    print("Duplicate date not found")
```


     

### Step 8. Ops...it seems the index is from the most recent date. Make the first entry the oldest date.




```python
apple = apple.sort_index(ascending=True)
apple
```


     

### Step 9. Get the last business day of each month.



```python
last  =  pd.date_range('1980-12-12','2014-07-08',freq='BM')
for date in last :
    print(date)
```



     

### Step 10. What is the difference in days between the first day and the oldest


     


```python
(pd.to_datetime('2014-07-08') - pd.to_datetime('1980-12-12')).days
```


### Step 11. How many months in the data we have?


     



```python
len(pd.date_range('1980-12-12','2014-07-08',freq='M'))
```



     

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches



```python
plt.figure(figsize=(13.5, 9))
plt.title('Adj Close Value Over Time')
plt.plot(apple['Date'], apple['Adj Close'], label='Adj Close')
plt.xlabel('Date')
plt.ylabel('Adj Close')
plt.show()
```
